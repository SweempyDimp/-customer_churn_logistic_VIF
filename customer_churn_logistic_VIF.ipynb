```python
for col in df.columns:
  if (df[col].dtype !='object'):
    sns.boxplot(df[col])
    plt.xlabel(col)
    plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder

le=LabelEncoder()

for col in df.columns:
  if df[col].dtype=='object':
    df[col]=le.fit_transform(df[col])
    for code, category in enumerate(le.classes_):
      print(f"{category}-> {code}")
    print()
```

```python
x=df.drop('Churn',axis=1)
y=df['Churn']
```

```python
vif_data=pd.DataFrame()
```

```python
vif_data
```

```python
vif_data['Features']=x.columns
```

```python
x.values
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

vif_values=[]
for i in range(len(x.columns)):
  vif=variance_inflation_factor(x.values,i) #calculating the vif value
  vif_values.append(vif)

vif_data['VIF_Values']=vif_values

vif_data
```

```python
x.drop('MonthlyCharges',axis=1,inplace=True)
```

```python
vif_data=pd.DataFrame()
vif_data['Features']=x.columns

vif_values=[]
for i in range(len(x.columns)):
  vif=variance_inflation_factor(x.values,i) #calculating the vif value
  vif_values.append(vif)

vif_data['VIF_Values']=vif_values

vif_data
```

```python
x.drop('PhoneService',axis=1,inplace=True)

vif_data=pd.DataFrame()
vif_data['Features']=x.columns

vif_values=[]
for i in range(len(x.columns)):
  vif=variance_inflation_factor(x.values,i) #calculating the vif value
  vif_values.append(vif)

vif_data['VIF_Values']=vif_values

vif_data
```

```python
from sklearn.model_selection import train_test_split
```
